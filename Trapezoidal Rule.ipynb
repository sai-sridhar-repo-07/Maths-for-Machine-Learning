```python
import matplotlib.pyplot as plt
```

```python
def trapezoidal_rule(f, a, b, n):

    # Define the partition
    partition = [a + i*(b-a)/n for i in range(n+1)]

    # Evaluate the function on the partition
    vals = [f(x) for x in partition]

    # Apply trapezoidal formula
    I_n = (b-a)/(2*n) * (vals[0] + vals[-1]) + (b-a)/n * sum(vals[1:-1])

    return I_n
```

```python
with plt.style.context("seaborn-v0_8"):
    plt.figure()

    ns = range(1, 25)  # n values
    Is = [trapezoidal_rule(lambda x: x**2, 0, 1, n) for n in ns]

    # Plot the true integral as a horizontal line
    plt.axhline(y=1/3, color='red', label="true integral (1/3)")

    # Plot the approximations
    plt.scatter(ns, Is, color='blue', label="trapezoidal_rule(f, a, b, n)")

    plt.ylim([0.3, 0.52])
    plt.xlabel("n (number of trapezoids)")
    plt.ylabel("Approximation")
    plt.title("Convergence of the Trapezoidal Rule")
    plt.legend()
    plt.show()
```
