```python
import numpy as np

def square(x):
    return x*x
```

```python
def cube(x):
    return x*x*x
```

```python
# composition of functions 

def compose(f,g):

    def composition(*args, **kwargs):
        return f(g(*args,**kwargs))

    return composition
```

```python
square_cube_composition = compose(square, cube)
square_cube_composition(2)  
```

```python
# functions as callable objects 
class Linear:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def __call__(self, x):
        return self.a * x + self.b
        
    def parameters(self):
        return {"a": self.a, "b": self.b}

    def __repr__(self):
        return f"Linear(a={self.a}, b={self.b})"

```

```python
f = Linear(2, -1)
f(2.1)
```

```python
f.parameters()
```

```python
# Base class for fucntion

class Function:
    def __init__(self):
        pass

    def __call__(self, *args, **kwargs):
        pass

    def parameters(self):
        return dict()
```

```python
class Sigmoid(Function):
    def __call__(self, x):
        return 1 / (1 + np.exp(-x))
```

```python
sigmoid = Sigmoid()
sigmoid(2)   

```

```python
sigmoid.parameters() 
```

```python
# Composition in object oriented way 
composed = compose(Linear(2,-1), Sigmoid())
composed(2)
```

```python
isinstance(composed, Function) 
```

```python
composed.parameters()  
```

```python
# Implementing OO composition

class Composition(Function):
    def __init__(self, *functions):
        self.functions = functions

    def __call__(self, x):
        for f in reversed(self.functions):
            x = f(x)
        return x
```

```python
composed = Composition(Linear(2,-1), Sigmoid())
composed(2)  
```

```python
isinstance(composed, Function) 
```

```python
composed.parameters()  
```
